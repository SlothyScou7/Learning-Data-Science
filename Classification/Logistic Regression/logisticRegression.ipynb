# Splitting the dataset

```python
X_discrete = df.drop(['Purchased'],axis=1)
y_discrete = df.Purchased
print(X_discrete.head())
```

```python
X_con = df_con.drop(['Purchased'],axis=1)
y_con = df_con.Purchased
X_con = X_con.drop(['User ID'],axis=1)
print(X_con.head())
```

```python
print(y_con.head())
```

# Training & Testing set of data for continuous

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_con,y_con,test_size=0.25,random_state=0)
```

# Feature scaling

```python
'''
why do we need to scale features ?
we need to scale features for better results 
In course the features are scaled let's try both with and without features..
'''
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='liblinear')
classifier.fit(X_train,y_train)
```

# Predicting the results

```python
print(classifier.predict(sc.transform([[19,76000,0,1]])))
```

```python
y_pred = classifier.predict(X_test)
y_pred = y_pred.reshape((len(y_pred),1))
y_test = y_test.reshape((len(y_pred),1))
x = np.concatenate((y_pred,y_test),axis=1)
print(x)
```

# Confusion Matrix

```python
'''
Compute confusion matrix to evaluate the accuracy of a classification.

By definition a confusion matrix  is such that  is equal to the number of observations known to be in group  and predicted to be in group .

Thus in binary classification, the count of true negatives is , false negatives is , true positives is  and false positives is .
'''
# it is used to calculate the accuracy of the classification model 
# it compares the 2d arrays column wise i think and make acuuracy table
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test,y_pred))
accuracy_score(y_test, y_pred)
'''
confusion matrix --->>
true_negative | False_positive
false_neg | true_pos


or
            predicted no      predicted yes
actual no 
actual yes 

'''
```

# making our own roc curve 

```python
# print(len(x))
count_negative = 0
true_negative = 0
negative = []
for i in range(0,len(x)):
    if x[i][0] != x[i][1]:
        count_negative+=1
        print(x[i][0],[x[i][1]])
        negative.append(i)
    if x[i][0] == x[i][1] and x[i][0]==0:
        true_negative +=1
true_positive = 91-true_negative
print("true_negative",true_negative)
print("true_positive:",true_positive) 
print(negative)
print("false positive  are:",6)
print("false negative:",3)
```

```python
'''

sensitivity  -->>It measures the fraction of the initial(total) positives which have been predicted correctly 
(or recalled) as positive by the classifier. 
'''
sensitivity = true_positive/(true_positive+3)
print(sensitivity)
```
