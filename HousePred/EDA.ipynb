## Correlation matrix

```python
# we would be plotting a correlation matrix 
```

```python
matrix = train.corr()
```

```python
train.corr()
```

```python
sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'})
f, ax = plt.subplots(figsize=(12,9))
sns.heatmap(matrix, vmax=.8, square=True)
```

```python
k = 10
cols = matrix.nlargest(k,'SalePrice')['SalePrice'].index # this returns the largest correlations with Sale Price
cols
```

```python
cm = np.corrcoef(train[cols].values.T)
f, ax = plt.subplots(figsize=(12,9))
sns.heatmap(cm,cbar=True,square=True,xticklabels=cols,annot=True,yticklabels=cols)
```
